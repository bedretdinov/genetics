```python
import numpy as np
from sklearn.metrics import mean_squared_error, accuracy_score
import pandas as pd
from random import randint
import random
from multiprocessing import Pool
from sklearn.model_selection import train_test_split


class GeneticAlgoritm:
    
    coef__ = None
    hist__ = []
    epoh__ = 1
    verbose__    = True
    selection__  = 2000
    population__ = 5000
    
    def __init__(self, epoh, selection=2000, population=5000, verbose=True):
        self.epoh__      = epoh
        self.selection__ = selection
        self.population__= population
        self.hist__      = []
        self.verbose__   = verbose
        
    def init_population(self, X):  
        population = np.array([ [ random.random()*10 for _ in range(X.shape[1])] for _ in range(self.population__)])
        return population
    
    def selection(self,population,  X,y):
        X = np.array(X)
        y_pred = [] 
        population_error = []
        
        for item in population:  
            item = np.array(item)
            
            try:
                y_pred = (item*X).sum(axis=1)
            except:
                print(item)
            
            mse = mean_squared_error(y, y_pred) 
            population_error.append([mse,item]) 
         
        population_error = pd.DataFrame(population_error).sort_values(0)
        take_effective = population_error.iloc[0:self.selection__]
        self.coef__ = take_effective[1].values[0]
        
        self.hist__.append(take_effective[0].values[0])
        
        if(self.verbose__):
            print(self.hist__[-1]) 
        
        return take_effective[1].values 
    
    def fit_paralel(self, params, core=1, workers=3):
        X,y = params  
        p = Pool(core*2+1)
        result = p.map(self.fit, [(X,y) for _ in range(workers)])
        p.close()
        return result
    
    def fit(self, params):
        X,y = params
        
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
        
        population = self.init_population(X)
        for _ in range(self.epoh__):
            effective = self.selection(population, X_train,y_train) 
            population = self.transgression(effective)
        
        if(self.verbose__):
            print('acc',accuracy_score(np.round(X_test.dot(self.coef__)), y_test))
        
        return self.hist__,self.coef__,accuracy_score(np.round(X_test.dot(self.coef__)), y_test)
        
    def transgression(self, X):
        
        X_len    = len(X)-1
        X_weight = len(X[0])
        
        generated = []
        newPopulation = []
        
        item = 0
        while True:
            
            item+=1
            
            if(len(newPopulation)==self.population__):
                break 
            
            parents = []
            parents.append(X[randint(0,X_len)])
            parents.append(X[randint(0,X_len)])
            
            cahild = []
            
            if(item%100==0): 
                for i in range(X_weight): 
                    cahild.append(random.random()*10)
            else:
            
                for i in range(X_weight):  
                     cahild.append(parents[np.random.choice(2, 1, p=[0.5, 0.5])[0]][i])

                    
            newPopulation.append(cahild)
       
        return newPopulation
```
